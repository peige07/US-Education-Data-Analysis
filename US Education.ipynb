```python
#For this analysis we only want to compare data across states in a single year so we will choose the year 2000
ndf2000 = ndf[ndf['YEAR'] == 2000]

ndf2000.shape
```

```python
#Checking for outliers in TOTAL_REVENUE

ndf2000['TOTAL_REVENUE'].describe()
```

```python
plt.figure(figsize=(5,5))
plt.boxplot(ndf2000['TOTAL_REVENUE'])
plt.show()
```

```python
#Quick check that our quartiles are correct
q3 = np.percentile(ndf2000['TOTAL_REVENUE'], 75)
q1 = np.percentile(ndf2000['TOTAL_REVENUE'],25)
print(f'Quartile number 3: {q3}')
print(f'Quartile number 1: {q1}')

iqr = q3 - q1

print(f'Interquartile range: {iqr}')

upper_limit = q3 + 1.5 * iqr
lower_limit = q1 -1.5 * iqr

print(f'Upper Limit: {upper_limit}')
print(f'Lower Limit: {lower_limit}')

# Identify outliers
outliers = (ndf2000['TOTAL_REVENUE'] < lower_limit) | (ndf2000['TOTAL_REVENUE'] > upper_limit)

# Display the outliers
print("Outliers:")

for key, value in outliers.items():
    if value == True:
        print(f'-{key}')
```

```python
#Ill use IQR to determine outliers but incase your would like to use Z-Scores ill provide the code as well
z_total_revenue = (ndf2000['TOTAL_REVENUE'] - ndf2000['TOTAL_REVENUE'].mean())/ndf2000['TOTAL_REVENUE'].std()

outliers = (z_total_revenue > 3) | (z_total_revenue< -3)

for key,value in outliers.items():
    if value == True:
        print(f'Row #{key}, is an outlier')
```

```python
#We have 3 outliers present so I will check to see which states they are

revenue_outliers = [412, 440, 451]

for row_index in revenue_outliers:
    print(f"Outliers in TOTAL_REVENUE: {ndf2000.loc[row_index, 'PRIMARY_KEY']}")
```

<h4>Hypothesis Testing</h4>

</div>

<div style="background-color:lightyellow; color:black; padding:20px;">
<p>H<sub>0</sub>: Total Revenue has no statistical affect on Average 8th Grade Math Scores.</p>
    <br>
<p>H<sub>1</sub>sub>: Total Revenue has a statistically significant affect Average 8th Grade Math Scores.</p>
</div>

```python
column1 = ndf2000['TOTAL_REVENUE']
column2 = ndf2000['AVG_MATH_8_SCORE']

correlation_coefficient, p_value = sci.stats.pearsonr(column1, column2)

print(f"Pearson correlation coefficient: {correlation_coefficient}")
print(f"P-value: {p_value}")

if p_value < 0.05:  
    print("The correlation is statistically significant.")
else:
    print("The correlation is not statistically significant.")
```

<div style="background-color:lightyellow; color:black; padding:20px;">
Based on the Pearson Correlation Coeffficient we have failed to prove our alternative hypothesis.
</div>

<hr>

<div style="background-color:lightyellow; color:black; padding:20px;">
<p>H<sub>0</sub>: Total Expenditure has no statistical affect on Average 8th Grade Math Scores.</p>
    <br>
<p>H<sub>1</sub>sub>: Total Expenditure has a statistically significant affect Average 8th Grade Math Scores.</p>
</div>

```python
column1 = ndf2000['TOTAL_EXPENDITURE']
column2 = ndf2000['AVG_MATH_8_SCORE']

correlation_coefficient, p_value = sci.stats.pearsonr(column1, column2)

print(f"Pearson correlation coefficient: {correlation_coefficient}")
print(f"P-value: {p_value}")

if p_value < 0.05:  
    print("The correlation is statistically significant.")
else:
    print("The correlation is not statistically significant.")
```

<div style="background-color:lightyellow; color:black; padding:20px;">
While our P-value was slightly higher for TOTAL EXPENDITURE we failed to prove our alternative hypothesis.
</div>
